```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic=pd.read_csv("train.csv")
```

```python
titanic.head(5)
```

```python
titanic.info()
```

```python
titanic.isnull().sum()
```

```python
sns.countplot(data=titanic,x='Sex',hue='Survived')
```

```python
numeric_cols=titanic.select_dtypes(include=['float64','int64']).columns
corr_mat=titanic[numeric_cols].corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_mat,annot=True,cmap='coolwarm',linewidths=1,linecolor='black')
plt.title('Correlation matrix of numeric features')
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.histplot(data=titanic,x='Fare',hue='Survived',kde=True)
plt.show()
```
